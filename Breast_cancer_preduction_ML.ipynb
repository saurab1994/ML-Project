Building Machine learning model for our dataset.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
x=df.iloc[:,:-1]
y =df['target']
X_train,X_test,y_train,y_test =train_test_split(x,y,random_state=2,test_size =.3)
```

```python
# logistic regression 
from sklearn.linear_model import LogisticRegression
pipe =make_pipeline(StandardScaler(),LogisticRegression())
param ={'logisticregression__C':[0.01,0.1,1,10,100]}
grid =GridSearchCV(pipe,param_grid=param,cv =5)
grid.fit(X_train,y_train)
print(f'beat_estimator :{grid.best_estimator_}')
print(' ')
print(f'training score of Logistic Regression : {grid.score(X_train,y_train)}')
print(f'testing score of Logistic Regression : {grid.score(X_test,y_test)}')
```

```python
# SVM 
from sklearn.svm import SVC
pipe =make_pipeline(StandardScaler(),SVC())
param ={'svc__C':[0.01,0.1,1,10,100],'svc__gamma':[0.01,0.1,1,10,100]}
grid =GridSearchCV(pipe,param_grid=param,cv =5)
grid.fit(X_train,y_train)
print(f'beat_estimator :{grid.best_estimator_}')
print(' ')
print(f'training score of support vector machine : {grid.score(X_train,y_train)}')
print(f'testing score of support vector machine : {grid.score(X_test,y_test)}')
```

```python
# Random Foresr clasiifier 
from sklearn.ensemble import RandomForestClassifier
pipe =make_pipeline(RandomForestClassifier())
param ={'randomforestclassifier__n_estimators':[100,150,200,250,300]}
grid =GridSearchCV(pipe,param_grid=param,cv =5)
grid.fit(X_train,y_train)
print(f'beat_estimator :{grid.best_estimator_}')
print(' ')
print(f'training score of Random Forest Classifier : {grid.score(X_train,y_train)}')
print(f'testing score of Random Forest Classifier : {grid.score(X_test,y_test)}')
```
